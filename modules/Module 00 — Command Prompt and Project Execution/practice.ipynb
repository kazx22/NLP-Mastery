## Section 2 — `dir`

## Section 6 — `del`, `rmdir`, `copy`, `move`

CMD commands:

```cmd
del file.txt
rmdir empty_folder
copy file.txt backup\
move file.txt backup\
```

Below, Python simulates creating, copying, moving, and deleting files.

````python
from pathlib import Path
import shutil

workspace = Path("cmd_practice_workspace")
workspace.mkdir(exist_ok=True)

original = workspace / "sample.txt"
original.write_text("Hello from a practice file.", encoding="utf-8")

backup_dir = workspace / "backup"
backup_dir.mkdir(exist_ok=True)

copied = backup_dir / "sample.txt"
shutil.copy2(original, copied)

moved = workspace / "moved_sample.txt"
shutil.move(str(original), str(moved))

print("Copied file exists:", copied.exists())
print("Moved file exists:", moved.exists())
print("Original file still exists after move:", original.exists())
````

## Section 10 — `pip install`, `pip list`, `pip freeze`

CMD commands:

```cmd
pip install pandas
pip list
pip freeze > requirements.txt
```

Below, we simulate what each command is for.

````python
examples = {
    "pip install pandas": "Install pandas into the current Python environment.",
    "pip list": "Show installed packages.",
    "pip freeze > requirements.txt": "Save exact package versions to requirements.txt.",
    "python -m pip install pandas": "Safer form because it uses the current Python interpreter."
}

for command, meaning in examples.items():
    print(command)
    print(" ->", meaning)
    print()
````

## Section 11 — Why `python -m src.main` matters

CMD command:

```cmd
python -m src.main
```

Meaning:
- run `main.py` as a module inside the `src` package

This is best for real projects.

Rule:
- simple file -> `python file.py`
- project file in `src/` -> `python -m src.file`

````python
examples = [
    ("python test.py", "Use for a simple standalone file."),
    ("python -m src.main", "Use for a project file inside src/."),
    ("python -m src.benchmark --model biobert", "Run a project module and pass an argument."),
]

for cmd, note in examples:
    print(cmd)
    print(" ->", note)
    print()
````
